```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn as skn
from scipy import stats
from scipy.stats import norm
```

```python
df = pd.read_csv('cardio_train.csv', sep=';')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
matrice_corr = df.corr()
f, ax = plt.subplots(figsize =(9, 8)) 
sns.heatmap(matrice_corr, ax = ax, cmap ="Paired", linewidths = 0.1, annot=True)
```

```python
sns.set_theme(style="ticks")
g = sns.catplot(x="height", y="weight", hue="gender", data=df,height=8, aspect=11/8)
g.set_xticklabels(rotation=45)
plt.legend()
```

```python
sns.set_theme(style="ticks")
g = sns.catplot(x="ap_hi", y="ap_lo", hue="cardio", data=df,height=8, aspect=11/8)
g.set_xticklabels(rotation=45)
plt.legend()
```

```python
sns.set_theme(style="ticks")
g = sns.catplot(x="ap_lo", y="ap_hi", hue="cardio", data=df,height=8, aspect=11/8)
g.set_xticklabels(rotation=45)
plt.legend()
```
